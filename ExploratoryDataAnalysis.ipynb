## Exploratory Data Analysis
#### Getting familiar with the data.
This is notebook is dedicated to EDA of the dataset. This is a more unconventional form of EDA since we are working with images as our data.

```python
%matplotlib inline

import os

import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 8})
```

Load in the labels.csv file as a pandas Dataframe
This dataframe will contain the ID of each dog with its corresponding breed

```python
breed_annotations = pd.read_csv("labels.csv")
breed_annotations.head()
```

```python
breed_annotations.describe()
```

We can see that all IDs are unique and that there are 120 different breeds.

Let's count the number of times a dog breed appears in this data set

```python
breed_count = breed_annotations["breed"].value_counts()

max_breed_type, max_breed_count = (breed_count.index[0], breed_count[0])
min_breed_type, min_breed_count = (breed_count.index[-1], breed_count[-1])

print(f"The {max_breed_type} appears {max_breed_count} time in this data set")
print(f"The {min_breed_type} appears {min_breed_count} times in this data set")
```

As we can see there is a large variance between the number of times a dog breed may appear.
It may be useful to visualize.

```python
count_plot = plt.figure(figsize=(10,20))
axes = count_plot.add_subplot(1,1,1)
axes.set_ylabel("Dog Breeds")
axes.set_xlabel("Count")
axes.set_title("Counts of breed in Labels dataset")
axes.barh(breed_count.index,breed_count)
```

As the bar plot shows, some breeds are over-represented in this dataset. One technique would be to compile our Neural Network using the Focal Loss function instead of Cross Entropy Loss. Focal Loss is excellent at dealing with an asymmetrical image dataset. Another technique would be to limit each class to 65 images per class and stick with the Cross Entropy Loss function. Both are an option.
